```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.layouts import layout
import pandas as pd

output_notebook()
```

```python
ch4 = pd.read_csv("../dados/dc_ch4_hol_fl02.csv", sep="\t")
co2 = pd.read_csv("../dados/dc_co2_hol_fl02.csv", sep="\t")
n2o = pd.read_csv("../dados/dc_n2o_hol_fl02.csv", sep="\t")

for name, i in zip(["ch4", "co2", "n2o"],[ch4, co2, n2o]):
    print(name, end=":")
    for col in i.columns:
        print(col, end=" | ")
    print()
```

```python
ch4_fig = figure(plot_width=450, plot_height=300,
           x_axis_label = "years BP", y_axis_label = "ppbv",
           title="CH4 concentration")

ch4_fig.line(ch4["age [yr BP]"], ch4["CH4 concentration [ppbv]"])
#ch4_fig.line(ch4["age [yr BP]"], ch4["one standard deviation [ppbv]"], color="gray")

co2_fig = figure(plot_width=450, plot_height=300,
           x_axis_label = "years BP", y_axis_label = "ppmv",
           title="CO2 concentration")

co2_fig.line(co2["age [yr BP]"], co2["CO2 concentration [ppmv]"], color="black")
#co2_fig.line(co2["age [yr BP]"], co2["one standard deviation [ppmv]"], color="gray")

n2o_fig = figure(plot_width=450, plot_height=300,
           x_axis_label = "years BP", y_axis_label = "ppbv",
           title="N2O concentration")

n2o_fig.line(n2o["age [yr BP]"], n2o["N2O concentration [ppbv]"], color="red")
#n2o_fig.line(n2o["age [yr BP]"], n2o["one standard deviation [ppbv]"], color="gray")

l = layout([ch4_fig, co2_fig], [n2o_fig])
show(l)
```
