```python
from bokeh.plotting import *
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, DataRange1d, PanTool, WheelZoomTool, BoxSelectTool
)
import pandas as pd

output_notebook()
```

```python
data = pd.read_csv("elnino.csv")
for i in enumerate(data.columns):
    print(i)
```

```python
sea_temp = data.Sea_Surface_Temp[-200:-1]
air_temp = data.Air_Temp[-200:-1]

fig_temp = figure(width=900, height=500)

fig_temp.line(range(len(sea_temp)),sea_temp, color="navy", legend="Sea Surface Temp")
fig_temp.line(range(len(air_temp)),air_temp, color="red", legend="Ait Temp")

show(fig_temp)
```

```python
humidity = data.Humidity[-200:-1]

fig_humidy = figure(width=900, height=500)

fig_humidy.line(range(len(humidity)),sea_temp, color="darkcyan", legend="humidity")

show(fig_humidy)
```

```python
z_wind = data.Zonal_Winds[-200:-1]
m_wind = data.Meridional_Winds[-200:-1]

fig_winds = figure(width=900, height=500)

fig_winds.line(range(len(z_wind)),z_wind, color="green", legend="Zonal Winds")
fig_winds.line(range(len(m_wind)),m_wind, color="seagreen", legend="Meridional Winds")

show(fig_winds)
```

```python
longs, lats = [], []

for long, lat in zip(data.Longitude[-200:-1], data.Latitude[-200:-1]):
#     if long not in longs:
    longs.append(long)
    
#     if lat not in lats:
    lats.append(lat)
```

```python
map_options = GMapOptions(lat=lats[3], lng=longs[1] ,map_type="roadmap", zoom=11)


plot = GMapPlot(
    x_range=DataRange1d(), y_range=DataRange1d(), map_options=map_options
)

source = ColumnDataSource(
    data=dict(
        lat=lats,
        lon=longs,
    )
)

circle = Circle(x="lon", y="lat", size=15, fill_color="blue", fill_alpha=0.8, line_color=None)
plot.add_glyph(source, circle)

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())

show(plot)
```
