```python
from bokeh.plotting import figure, output_notebook, show
import pandas as pd

output_notebook()
```

```python
data = pd.read_csv("../dados/cdc_zika.csv")
```

```python
data.dtypes
```

```python
PE = data[(data.location=="Brazil-Pernambuco")]

qnt_conf = PE[data.data_field=="microcephaly_confirmed"].value
qnt_investig = PE[data.data_field=="microcephaly_under_investigation"].value
report_date = PE[data.data_field=="microcephaly_confirmed"].report_date
```

```python
p = figure(plot_width=800, plot_height=400,
           x_axis_label="Data", y_axis_label="qnt",
           x_range=report_date.tolist(),
           title="confirmados-PE")

p.line(report_date, qnt_conf, color="red", legend="confirmados")
p.line(report_date, qnt_investig, legend="sob investigação")

show(p)
```

```python
YPL = data[(data.location=="Colombia-Casanare-Yopal")]

suspeito = YPL[data.data_field=="zika_suspected_clinic"].value
confirmado = YPL[data.data_field=="zika_confirmed_laboratory"].value
report_date_col = YPL[data.data_field=="zika_confirmed_laboratory"].report_date
```

```python
p2 = figure(plot_width=800, plot_height=400,
           x_axis_label="Data", y_axis_label="qnt",
           x_range=report_date.tolist(),
           title="Colombia-Casanare-Yopal")

p2.line(report_date_col, suspeito, legend="suspeito")
p2.line(report_date_col, confirmado, color="red", legend="confirmado")

show(p2)
```

```python
p3 = figure(plot_width=800, plot_height=400,
           x_axis_label="Data", y_axis_label="qnt",
           x_range=report_date.tolist(),
           title="cofirmados: Brasil-Pernambuco - Colombia-Casanare-Yopal")

p3.line(report_date_col, qnt_conf, color="black",legend="BR-PE")
p3.line(report_date_col, confirmado, color="red", legend="COL-CSN-YPL")

show(p3)
```
