```python
from bokeh.plotting import output_notebook, show, figure
from bokeh.layouts import layout
import pandas as pd

output_notebook()
```

```python
data = pd.read_csv("../dados/historico_tempo_curado.csv", sep=";")
data.columns
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.describe()
```

```python
dias_prec = data.NumDiasPrecipitacao
prec_total = data.PrecipitacaoTotal
temp_max = data.TempMaximaMedia
temp_min = data.TempMinimaMedia
periodo = data.Data

p1 = figure(plot_width=450, plot_height=300,
            x_range=periodo.tolist(), title="precipitacao - total")
p1.line(periodo, prec_total, color="red")

p2 = figure(plot_width=450, plot_height=300,
            x_range=periodo.tolist(), title="precipitacao - dias")
p2.line(periodo, dias_prec, color="darkcyan")

p3 = figure(plot_width=450, plot_height=300,
            x_range=periodo.tolist(), title="temp. max")
p3.line(periodo, temp_max, color="darkorange")

p4 = figure(plot_width=450, plot_height=300,
            x_range=periodo.tolist(), title="temp. min")
p4.line(periodo, temp_min, color="navy")

l = layout([p1, p2], [p3, p4])
show(l)
```
