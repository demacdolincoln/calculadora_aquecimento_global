```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
import csv
output_notebook()
```

```python
data = pd.read_csv("../dados/co2_por_nacao.csv")
data.columns
```

```python
x = "Year"
y = "Per capita CO2 emissions (metric tons of carbon)"
```

```python
brasil = data[data.Nation=="BRAZIL"][[x, y]]
```

```python
p = figure(plot_width=500, plot_height=300, x_axis_label = x, title=y)
p.scatter(brasil[x], brasil[y])
show(p)
```

```python
from sklearn.svm import SVR
```

```python
br_data = brasil[brasil.Year>1949]
X = br_data[x].values.reshape(-1, 1)
Y = br_data[y].values.reshape(-1, 1)
```

```python
reg = SVR(kernel="linear")
results = reg.fit(X, Y).predict(X).reshape(-1,1)
```

```python
p2 = figure(plot_width=500, plot_height=300, x_axis_label = x, title=y)
p2.scatter(brasil[x], brasil[y])
p2.line(x=br_data[x], y=results.tolist(), line_width=2, line_color="darkorange")
show(p2)
```

```python
country = "JAMAICA"
pais = data[data.Nation==country][[x, y]]

p3 = figure(plot_width=500,
    plot_height=300,
    x_axis_label = x,
    title = country+"--"+y)

p3.scatter(brasil[x], brasil[y], color="indigo", legend="BRASIL")
p3.scatter(pais[x], pais[y], color="darkorange", legend=country)

p3.legend.location="bottom_right"

show(p3)
```
