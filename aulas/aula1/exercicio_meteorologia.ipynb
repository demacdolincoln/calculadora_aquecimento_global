```python
from bokeh.plotting import output_notebook, show, figure
from bokeh.layouts import layout
import pandas as pd

%matplotlib notebook
output_notebook()
```

```python
data_meteorologia = pd.read_csv("recife_052017.csv", sep=";", header=16)
```

```python
data_meteorologia.columns
```

```python
data_meteorologia.head()
```

```python
prec = data_meteorologia.Precipitacao[1:-1:2]
datas = data_meteorologia.Data[1:-1:2]

plot_prec = figure(plot_width=900, plot_height=300,
            x_range=datas.tolist(), title="Precipitação")

plot_prec.line(datas, prec, color="black")
show(plot_prec)
```

```python
maxi = data_meteorologia.TempMaxima[0:-1:2]
mini = data_meteorologia.TempMinima[1:-1:2]
datas = data_meteorologia.Data[1:-1:2]

plot_temp = figure(plot_width=900, plot_height=300,
            x_range=datas.tolist(), title="Temp. Máximas e Mínimas")

plot_temp.line(datas, maxi, color="darkred")
plot_temp.line(datas, mini, color="darkcyan")
show(plot_temp)
```

```python
veloc_vento = data_meteorologia["Velocidade do Vento Media"][0:-1:2]
datas = data_meteorologia.Data[0:-1:2]

plot_vento = figure(plot_width=900, plot_height=300,
            x_range=datas.tolist(), title="Velocidade do Vento Media")

plot_vento.line(datas, veloc_vento, color="green")
show(plot_vento)
```
