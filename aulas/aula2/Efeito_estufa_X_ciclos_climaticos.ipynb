```python
from bokeh.plotting import output_notebook, show, figure
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")

output_notebook()
```

```python
emissao_co2 = pd.read_csv("../../dados/co2_por_nacao.csv")
nivel_oceanico = pd.read_csv("../../dados/nivel_oceanico.csv", header=5)
```

## anotações iniciais

```python
emissao_co2.columns
```

```python
n = ""
for i in emissao_co2.Nation:
    if i != n:
        print(i)
    n = i
```

```python
indicador = 'Per capita CO2 emissions (metric tons of carbon)'

paises = ["COLOMBIA", "BRAZIL", "ARGENTINA", "CHILE", "URUGUAY", "VENEZUELA"]
colors = ["red", "green", "black", "blue", "cyan", "orange"]

p_co2 = figure(plot_width=900, plot_height=250, title=indicador)

for i, j in zip(paises, colors):
    pais = emissao_co2[emissao_co2.Nation==i]
    p_co2.line(pais["Year"],
               pais[indicador],
               color=j, legend=i)
    
p_co2.legend.orientation="horizontal"
show(p_co2)
```

```python
from sklearn.svm import SVR

x = "Year"
y = "Per capita CO2 emissions (metric tons of carbon)"

brasil = emissao_co2[emissao_co2.Nation=="BRAZIL"][[x, y]]

br_data = brasil[brasil.Year>1949]
X = br_data[x].values.reshape(-1, 1)
Y = br_data[y].values.reshape(-1, 1)

reg = SVR(kernel="linear")
results = reg.fit(X, Y).predict(X).reshape(-1,1)

p_br = figure(plot_width=500, plot_height=300, x_axis_label = x, title=y)
p_br.scatter(brasil[x], brasil[y])
p_br.line(x=br_data[x], y=results.tolist(), line_width=2, line_color="darkorange")
show(p_br)
```
