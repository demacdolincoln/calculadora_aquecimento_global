```python
from bokeh.plotting import output_notebook, show, figure
from bokeh.models import BoxAnnotation
import pandas as pd
from numpy import logical_and as l_and
from warnings import filterwarnings
filterwarnings("ignore")

output_notebook()
```

```python
emissao_co2 = pd.read_csv("../../dados/co2_por_nacao.csv")
```

```python
for i in enumerate(emissao_co2.columns):
    print(i)
```

```python
brasil = emissao_co2[emissao_co2.Nation == "BRAZIL"]
```

```python
plot_comb_foss = figure(plot_width=900, plot_height=450, title="EmissÃµes de co2 no Brasil")
```

```python
periodos = [1930, 1945, 1964, 1985]
periodos_colors = iter(["green","purple", "yellow", "black", "red"])

for p in range(len(periodos)):
    box = None
    if p == 0:
        box = BoxAnnotation(right=periodos[p], fill_color=next(periodos_colors))
    else:
        box = BoxAnnotation(left=periodos[p-1], right=periodos[p], fill_color=next(periodos_colors))
    plot_comb_foss.add_layout(box)

box = BoxAnnotation(left=periodos[-1], fill_color=next(periodos_colors))
plot_comb_foss.add_layout(box)
```

```python
combustiveis = [
    "Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)",
    "Emissions from solid fuel consumption",
    "Emissions from liquid fuel consumption",
    "Emissions from gas fuel consumption",
    "Emissions from gas flaring",
    "Emissions from bunker fuels (not included in the totals)"
]

colors = ["red", "black", "blue", "green", "darkorange", "purple"]

for item, cor in zip(combustiveis, colors):
    plot_comb_foss.line(brasil.Year, brasil[item], color=cor, legend=item, line_width=1.3)
    
plot_comb_foss.legend.location = (70, 420)
# plot_comb_foss.legend.background_fill_alpha = 0.5

show(plot_comb_foss)
```
